```python
train_set = pd.DataFrame(data=None, columns = dataset.columns)
test_set = pd.DataFrame(data=None, columns = dataset.columns)
#train_set
```

### Step-2

```python
def train_test_splitting(full_dataframe:pd.DataFrame, tgt_class:str):
    global train_set
    global test_set
    filtered_dataframe = full_dataframe[full_dataframe.letter==tgt_class]
    train, test = train_test_split(filtered_dataframe, test_size = 0.3)
    train_set = pd.concat([train_set,train], axis=0)
    test_set = pd.concat([test_set,test], axis=0)
```

### Step-3

```python
for tgt_class in dataset.letter.unique():
    #print("target class is:",tgt_class)
    train_test_splitting(dataset, tgt_class)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import svm
from sklearn import metrics
from collections import Counter
```

### As we can see (refer Q2) that the range of the all independent variables is between 0 and 15 so standardization is not required

```python
unique_letters = dataset['letter'].sort_values().unique()
```

```python
positive_class = []
negative_class = []
total_rows = 0
for positive in range(0,len(unique_letters)):
    for negative in range(positive + 1, len(unique_letters)):
        positive_class.append(unique_letters[positive])
        negative_class.append(unique_letters[negative])
        total_rows = total_rows + 1 
```

```python
print("The total number of rows should be 26c2: ", total_rows)
```

```python
accuracy_df2 = pd.DataFrame(positive_class, columns = ['Positive_Class'])
accuracy_df2['Negative_Class'] = negative_class
accuracy_df2['Test_Accuracy_Linear_C5'] = np.NaN
```

```python
accuracy_df2
```

### Startified sampling can be used for splitting the data set into 70% training and 30% test by each target class

```python
tgt_class = dataset['letter']
train, test = train_test_split(dataset, test_size=0.3, random_state=0,stratify=tgt_class)
train.head(2)
```

```python
#train['target'] = np.NaN
#train.head(2)
```

### Pair-wise SVM claiisifier for each class

```python
for i in range(0,len(accuracy_df2)):
    
    positive_class_iter = accuracy_df2.iloc[i,0]
    negative_class_iter = accuracy_df2.iloc[i,1]
    print("Sequence no. :",str(i)," Pair: ",positive_class_iter," - ",negative_class_iter,"\n")
    
    train_subset = train[(train['letter'] == positive_class_iter) | (train['letter'] == negative_class_iter)]
    test_subset = test[(test['letter'] == positive_class_iter) | (test['letter'] == negative_class_iter)]
 
      
    train_subset['target'] = np.where(train_subset['letter']==positive_class_iter, 1, 0)
    test_subset['target'] = np.where(test_subset['letter']==positive_class_iter, 1, 0)
    X_train = train_subset.drop(['letter','target'],axis=1)
    X_test = test_subset.drop(['letter','target'],axis=1)
    y_train = train_subset['target']
    y_test = test_subset['target']
    
    svm2_mod1 = svm.SVC(kernel='linear',C=5.0)
    svm2_mod1.fit(X_train,y_train)
    y_pred1 = svm2_mod1.predict(X_test)
    accu_mod1 = metrics.accuracy_score(y_test, y_pred1)
    
    accuracy_df2.iloc[i,2] = accu_mod1
```
